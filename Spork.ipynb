```python
import spork

repo = spork.Repo()
await repo.connect_websocket("wss://sync.automerge.org")
```

```python
doc = await repo.create()
```

```python
await doc.set_string("title", "Spork self")
```

```python
await doc.dump()
```

```python
await doc.get_string("title")
```
