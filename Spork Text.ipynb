Spork Text Ops

```python
import spork
repo = spork.Repo()
await repo.connect_websocket("wss://sync.automerge.org")
```

```python
doc = await repo.find("automerge:4N4WaWxXd37ySkoJ7TdxUijTLNZF")

if doc == None:
    raise Exception("Document not found")

await doc.get_string("title")
```

```python
text = await doc.put_text("content", "Hello World")
```

```python
text = await doc.get_text("content")
if text:
    content = await text.get()
```

```python
content
```

```python
await text.splice(6, 0, "Beautiful ")
```

```python
await text.splice(5, 6, "")
```

```python
await text.splice(0, 5, "Hi")
```

```python
await text.get()
```

```python
# Insert without deleting
await text.insert(5, "amazing ")

# Delete characters
await text.delete(5, 7)

# Append to end
await text.append(" more text")
```

```python
content = await text.get()
content
```
